```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_points(X,Y,xlabel,ylabel):
    plt.plot(X,Y,"o")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
print(label_encoder)
```

```python
df["O_Encoded"] = label_encoder.fit_transform(df["O"])
print(df["O_Encoded"])
```

```python
X = df[["C1","C2","C3","C4","C5","C6","C7","C8","C9","C10","C12","C13","C14","C15","C16","C17","C18","C19","C20","C21","C22","C23","C24","C25","C26","C27","C28","C29","C30","C31","C32","C33","C34","C35","C36","C37","C38","C39","C40","C41","C42","C43","C44","C45","C46","C47"
]]#Features
Y = df["O_Encoded"]#Target
print(X)
print(Y)
plot_points(X,Y,"x","y")
```

We split our data in two parts. One is training set which involves data fields upon which the ML model is trained and developed. The test set is used to test the accuracy of predictions made by the trained model.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
#test size = 0.2 denotes that 20% of the data will be in the test state. Random_state provides the seed value.
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
print(regressor)
regressor.fit(x_train,y_train)#Used to find the most accurate equation for our LR model as per the dataset
```

```python
print(regressor.intercept_)#value of b
print(regressor.coef_)#We will get 46 coefficients for 46 variables
```

```python
y_test_predict = regressor.predict(x_test)
rounded = np.round(y_test_predict)
y_test_predict_rounded = np.array([])
y_test_predict_rounded = np.append(y_test_predict_rounded,rounded)
print(y_test_predict_rounded,"\n")
indices = np.where(y_test_predict_rounded>11)
y_test_predict_rounded[indices] = np.random.randint(0, 12, size=len(indices[0]))
print(y_test_predict_rounded)
y_test_predict_rounded_int = y_test_predict_rounded.astype(int)
y_test_predict_rounded_decoded = label_encoder.inverse_transform(y_test_predict_rounded_int)
print(y_test_predict_rounded_decoded,"\n")
```
